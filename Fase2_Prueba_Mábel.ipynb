```python
import pandas as pd

# Leer el fichero csv desde GitHub
url = "https://raw.githubusercontent.com/Violainedlst/proyecto-da-promo-H-modulo-3-team-2-People-pulse/main/HR_RAW_DATA.csv"
data_bruto = pd.read_csv(url)

# Para ver el fichero entero
pd.set_option('display.max_columns', None)

# Crear una copia del DataFrame para no actuar directamente sobre el original
data = data_bruto.copy()
data.head()
```

```python
# Convertir los nombres de las columnas en minúsculas
data.columns = [column.lower().strip() for column in data.columns]

# Modificar los nombres de las columnas para mejor comprensión lectora añadiendo una '_'
data.rename(columns={
    'unnamed: 0' : 'id',
    'businesstravel': 'business_travel',
    'dailyrate': 'daily_rate',
    'distancefromhome': 'distance_from_home',
    'educationfield': 'education_field',
    'employeecount': 'employee_count',
    'employeenumber': 'employee_number',
    'environmentsatisfaction': 'environment_satisfaction',
    'hourlyrate': 'hourly_rate',
    'jobinvolvement': 'job_involvement',
    'joblevel': 'job_level',
    'jobrole': 'job_role',
    'jobsatisfaction': 'job_satisfaction',
    'maritalstatus': 'marital_status',
    'monthlyincome': 'monthly_income',
    'monthlyrate': 'monthly_rate',
    'numcompaniesworked': 'num_companies_worked',
    'over18': 'over_18',
    'overtime': 'overtime',
    'percentsalaryhike': 'percent_salary_hike',
    'performancerating': 'performance_rating',
    'relationshipsatisfaction': 'relationship_satisfaction',
    'standardhours': 'standard_hours',
    'stockoptionlevel': 'stock_option_level',
    'totalworkingyears': 'total_working_years',
    'trainingtimeslastyear': 'training_times_last_year',
    'worklifebalance': 'work_life_balance',
    'yearsatcompany': 'years_at_company',
    'yearsincurrentrole': 'years_in_current_role',
    'yearssincelastpromotion': 'years_since_last_promotion',
    'yearswithcurrmanager': 'years_with_curr_manager',
    'sameasmonthlyincome': 'same_as_monthly_income',
    'datebirth': 'year_birth',
    'roledepartament': 'role_department',
    'numberchildren': 'number_children',
    'remotework': 'remote_work'
}, inplace=True)

data.head()
```

```python
columnas_a_eliminar = ["id", "same_as_monthly_income", "employee_count", "salary", "standard_hours", "over_18", "performance_rating", "number_children", "years_in_current_role"]
columnas_existentes = [col for col in columnas_a_eliminar if col in data.columns]
data = data.drop(columns=columnas_existentes)
data.head()
```

```python
data = data.applymap(lambda x: x.strip() if isinstance(x, str) else x)
data.head()
```

```python
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = data[col].str.lower()
data.head()
```

```python
data = data.astype(str)
data.head()
```

```python
data['daily_rate'] = data['daily_rate'].str.replace('$', '', regex=False)
data['daily_rate'].unique()
```

```python
data['business_travel'] = data['business_travel'].str.replace('-', '_', regex=False)
data['business_travel'].unique()
```

```python
data['distance_from_home'] = data['distance_from_home'].str.replace('-', '', regex=False)
data['distance_from_home'].unique()
```

```python
data['hourly_rate'] = data['hourly_rate'].str.replace('not available', '0', regex=False)
data['hourly_rate'].unique()
```

```python
data['remote_work'] = data['remote_work'].apply(lambda x: 'yes' if x in ['1', 'true', 'yes'] else 'no')
data['remote_work'].unique()
```

```python
data["gender"] = data["gender"].replace({'0': 'male', '1': 'female'})
data['gender'].unique()
```

```python
data["marital_status"] = data["marital_status"].str.replace('marreid', 'married', regex=False)
data['marital_status'].unique()
```

```python
con_coma = ['daily_rate', 'monthly_income', 'employee_number', 'total_working_years', 'work_life_balance']
for col in con_coma:
    data[col] = data[col].str.replace(',', '.', regex=False)
data.head()
```

```python
data = data.fillna('n/a')
data = data.replace('nan', 'n/a')
data.head()
```

```python
reemplazos = {'forty-seven': '47', 'fifty-eight': '58', 'thirty-six': '36', 'fifty-five': '55', 'fifty-two': '52', 'thirty-one': '31', 'thirty': '30',
              'twenty-six': '26', 'thirty-seven': '37', 'thirty-two': '32', 'twenty-four': '24'}
data['age'] = data['age'].replace(reemplazos)
data['age'].unique()
```

```python
for column in data.columns:
    print(f"Columna: {column}")
    print(data[column].unique())
    print("\n--------------------\n")
```
